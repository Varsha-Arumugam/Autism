```python
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
M=tf.keras.models.load_model('autism_fin.h5')
```

```python
image_size = (180, 180)
```

```python
arr=["autism/consolidated/autistic/0001.jpg","autism/test/non_autistic/002.jpg","autism/test/non_autistic/007.jpg","autism/test/autistic/009.jpg"]
```

```python
for x in arr:
    img = keras.preprocessing.image.load_img(x, target_size=image_size)
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) 

    predictions = M.predict(img_array)
    score = predictions[0]
    img = mpimg.imread(x)
    imgplot = plt.imshow(img)
    plt.show()
    print("Autistic") if(score<0.5) else print("Non-autistic")
    print("This image is %.2f percent autistic and %.2f percent non-autistic."% (100 * (1 - score), 100 * score))
    
```
