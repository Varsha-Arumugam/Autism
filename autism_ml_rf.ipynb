```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from distutils.dir_util import copy_tree
```

```python
pathTrain=r'c:\Users\VARSHA\Desktop\clg\ML\autism\train'
pathValid=r'c:\Users\VARSHA\Desktop\clg\ML\autism\valid'
pathTest=r'c:\Users\VARSHA\Desktop\clg\ML\autism\test'
categories=['autistic','non_autistic']
train =[]
valid =[]
test=[]
```

```python
def prepareData(path):
    data =[]
    for category in categories:
        p=os.path.join(path, category)
        lable= categories.index(category) 
        for img in os.listdir(p):
            img_path = os.path.join(p,img)
            image= cv2.imread(img_path ,0)
            image= cv2.resize(image,(50,50))# size of image 50*50
            last_image = np.array(image).flatten()
            data.append([last_image,lable])
    print("size of data: in path: ",path," :::: ",len(data))
    return  data
```

```python
train =prepareData(pathTrain)
valid =prepareData(pathValid)
test=prepareData(pathTest)
```

```python
pick_in_train = open('train.pickle','wb')
pickle.dump(train,pick_in_train)
pick_in_train.close()

pick_in_valid = open('valid.pickle','wb')
pickle.dump(valid,pick_in_valid)
pick_in_valid.close()

pick_in_test = open('test.pickle','wb')
pickle.dump(test,pick_in_test)
pick_in_test.close()
```
